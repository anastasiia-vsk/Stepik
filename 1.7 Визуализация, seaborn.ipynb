```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sp = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
sp = sp.rename(columns = {'parental level of education': 'parental_level_of_education',
                        'test preparation course': 'test_preparation_course',
                        'math score': 'math_score',
                        'reading score': 'reading_score',
                        'writing score': 'writing_score'})
```

```python
sp
```

```python
sp.math_score.hist()
```

```python
sp.plot.scatter(x = 'math_score', y = 'writing_score')
```

```python
ax = sns.lmplot(x = 'math_score', y = 'writing_score', hue = 'gender', data = sp, fit_reg = False)
ax.set_xlabels('MATH SCORE')
ax.set_ylabels('WRITING SCORE')
```
